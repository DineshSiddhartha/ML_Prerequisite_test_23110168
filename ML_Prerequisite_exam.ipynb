```python
import random
import numpy as np
import time
n=10
k=15
m=12
lis1=[]
lis2=[]
for i in range(n):
    lis = []
    for _ in range(k):
        lis.append(random.randint(0, 10))
    lis1.append(lis)
for i in range(k):
    lis = []
    for _ in range(m):
        lis.append(random.randint(0, 10))
    lis2.append(lis)
start=time.time()
final_list=[]
for i in range(n):
    lis=[]
    for u in range(m):
        s=0
        for j in range(k):
            s+=lis1[i][j]*lis2[j][u]
        lis.append(s)
    final_list.append(lis)
end=time.time()
for i in final_list:
    print(i)
print('time taken for execution of python code=',end-start)

start=time.time()
final_lis=np.dot(lis1,lis2)
end=time.time()
for i in final_lis:
    print(i)
print('time taken for execution of numpy code=',end-start)
```

What is the gradient of the following function with respect to x and y?
$$
x^2 \cdot y + y^3 \cdot sin(x)
$$

$$
\frac{\partial (x^2 \cdot y + y^3 \cdot sin(x))}{\partial x} 
= 2x \cdot y + x^2 \cdot (\frac{\partial y}{\partial x}) + 3 \cdot y^2 \cdot (\frac{\partial y}{\partial x}) \cdot sin(x) + y^3 \cdot cos(x)
$$
$$
= 2x \cdot y + y^3 \cdot cos(x) =u(x,y)
$$
$$
\frac{\partial (x^2 \cdot y + y^3 \cdot sin(x))}{\partial y} 

= 2x \cdot (\frac{\partial x}{\partial y}) \cdot y + x^2 + 3 \cdot y^2 \cdot sin(x) + y^3 \cdot cos(x) \cdot (\frac{\partial x}{\partial y})
$$
$$
= x^2 + 3 \cdot y^2 \cdot sin(x) =v(x,y)
$$

$$
u(0,2)=8
$$
$$
v(0,2)=0
$$
$$
u(3.14,2)=4.56
$$
$$
v(3.14,2)=9.86
$$

```python
import jax
import jax.numpy as jnp
def f(x,y):
    return x**2*y+y**3*jnp.sin(x)
gradient=jax.grad(f,0)
result=gradient(0.0,2.0)
print(result)

result=gradient(3.14,2.0)
print(result)
```
